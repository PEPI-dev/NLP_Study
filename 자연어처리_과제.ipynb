# BPE(Byte Pair Encoding)
- 텍스트 압축 및 자연어 처리에서 사용되는 특별한 토큰화 기술입니다. BPE는 기본적으로 텍스트에서 가장 빈번하게 나타나는 바이트 쌍을 찾아내고 이를 하나의 새로운 토큰으로 대체하여 텍스트를 압축하는 방법을 사용합니다.

### 동작 과정
1. 초기화: 각 문자를 하나의 토큰으로 간주합니다.

2. 빈도 계산: 주어진 텍스트에서 모든 바이트 쌍의 빈도를 계산합니다.

3. 가장 빈번한 바이트 쌍 찾기: 가장 빈번하게 등장하는 바이트 쌍을 찾아내어 하나의 새로운 토큰으로 대체합니다.

4. 토큰 업데이트: 찾아낸 새로운 토큰을 기존의 토큰들과 결합합니다.

5. 반복: 일정한 횟수나 더 이상의 특정 조건을 만족할 때까지 3-4단계를 반복합니다.

# WordPiece Tokenizer
- 자연어 처리에서 사용되는 토큰화 기술 중 하나로, 단어를 여러 서브워드(subword)로 분리하는 알고리즘이다. 이 기술은 자주 사용되는 단어나 구문을 하나의 토큰으로 만들고, 그 외의 단어는 서브워드로 나누어 처리함으로써 언어의 다양성을 표현할 수 있다.

## 주요 특징
- 훈련 데이터에 기반한 학습: WordPiece Tokenizer는 초기에 학습 데이터를 기반으로 하여 토큰화 알고리즘을 학습합니다. 주어진 텍스트에서 서브워드를 추출하고, 빈도에 따라 토큰을 생성하는 방식으로 동작합니다.

- Greedy 기반 분리: WordPiece Tokenizer는 탐욕적인(Greedy) 접근 방식을 사용하여 주어진 텍스트를 서브워드로 분리합니다. 빈도가 가장 높은 서브워드를 선택하고, 텍스트에서 해당 서브워드를 찾아 새로운 토큰으로 대체합니다. 이 작업을 반복하여 최종적으로 분리된 토큰들을 얻습니다.

- 언어의 다양성 처리: WordPiece Tokenizer는 자주 사용되는 단어나 어구를 하나의 토큰으로 처리함으로써 언어의 다양성을 유지하면서도 효과적으로 토큰화를 수행할 수 있습니다. 이는 흔한 용어나 구문을 하나의 단어로 처리하면서 미등록 단어(Out-of-Vocabulary, OOV)에 대한 대처도 가능하게 합니다.

- 순차적 학습을 통한 동적 업데이트: 학습 데이터에 따라 토큰화 알고리즘이 동적으로 변화할 수 있습니다. 새로운 데이터가 들어올 때마다 토큰화 알고리즘이 업데이트될 수 있어, 다양한 도메인에 적용이 가능합니다.
