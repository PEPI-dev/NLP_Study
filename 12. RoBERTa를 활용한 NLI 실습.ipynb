```python
#pipeline : 자연어 처리 작업을 간단한 코드로 여러 작업을 한버ㄴ에 저워하ㅣ
# AutoTokenizer : 자동으로 적절한 토크나이저를 선택하여 모델을 토큰화 할 수 있ㅇ므

from transformers import pipeline,AutoTokenizer
```

```python
classifier = pipeline(
  'text-classification',
  model = 'Huffon/klue-roberta-base-nli',
  return_all_scores=True
)
```

```python
tokenizer = AutoTokenizer.from_pretrained('Huffon/klue-roberta-base-nli')
tokenizer
```

```python
tokenizer.sep_token
```

```python
classifier(f'나는 게임을 한느 것을 좋아해 {tokenizer.sep_token} 나는 게임을 하는 것이 싫어')
```

```python
# ENTAILMENT : 첫 번쨰 문장이 두번쨰 문장을 내포하거나 함축한 경우
# NEUTRAL : 두 문장의 관계가 특별히 긍정적이나 부정적이지 않을 떄(모순 관계가 없을 때)
# CONTRADICTION : 모순 관계가 있을 때
classifier(f'나는 게임을 하는 것을 좋아해{tokenizer.sep_token} 나는 게임을 하는 것')
```

```python
classifier(f'여러 남성들이 축구를 하고 있어요{tokenizer.sep_token} 어떤 남자들은 스포츠를 하고 있어요')
```
